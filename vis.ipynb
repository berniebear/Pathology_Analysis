```python
%matplotlib inline
import os, sys
import numpy as np
import csv
import matplotlib
import matplotlib.pyplot as plt
```

# Train from scratch 

```python
# Train from scratch
log_dir = os.path.join('resnet', 'checkpoint')
accs = []
for i in range(4):
    acc = []
    log_file = os.path.join(log_dir, 'log_Res18_sp{}_no.txt'.format(i))
    with open(log_file) as lines:
        for line in lines:
            ep, a = line.strip().split(',')
            acc.append(float(a))
    accs.append(acc)
    if len(acc) > 1:
        print(np.max(np.asarray(acc)))
accs=np.asarray(accs)
# 96.1616161616
# 96.4646464646
# 96.2626262626
# 95.2525252525
```

```python
for i in range(4):
    plt.plot(accs[i])
plt.ylim([80,100])
plt.grid()
plt.show()
maxs=np.max(accs,axis=1)
print('logs: {}'.format(maxs))
print('mean: {}'.format(np.mean(maxs)))
print('std: {}'.format(np.std(maxs)))
# with augmentation
#logs: [96.16161616 96.46464646 96.26262626 95.25252525]
#mean: 96.0353535353
#std: 0.464948298973
# without augmentation
#logs: [94.84848485 96.16161616 95.45454545 95.65656566]
#mean: 95.5303030303
#std: 0.470402424493
```

# Transfer Learning (pre-trained on cifar10) 

```python
# Train from scratch
log_dir = os.path.join('resnet', 'checkpoint')
accs = []
for i in range(4):
    acc = []
    log_file = os.path.join(log_dir, 'log_Res18_trans_sp{}_no.txt'.format(i))
    with open(log_file) as lines:
        for line in lines:
            ep, a = line.strip().split(',')
            acc.append(float(a))
    accs.append(acc)
    if len(acc) > 1:
        print(np.max(np.asarray(acc)))
accs=np.asarray(accs)
# 97.2727272727
# 96.3636363636
# 97.1717171717
# 96.9696969697
```

```python
for i in range(4):
    plt.plot(accs[i])
plt.ylim([80,100])
plt.grid()
plt.show()
maxs=np.max(accs,axis=1)
print('logs: {}'.format(maxs))
print('mean: {}'.format(np.mean(maxs)))
print('std: {}'.format(np.std(maxs)))
# with augmentation
#logs: [97.27272727 96.36363636 97.17171717 96.96969697]
#mean: 96.9444444444
#std: 0.352632324344
# no augmentation
#logs: [96.26262626 97.17171717 96.76767677 96.96969697]
#mean: 96.7929292929
#std: 0.337855761628
```
