```python
%matplotlib inline
import os, sys
import numpy as np
import csv
import matplotlib
import matplotlib.pyplot as plt
```

# Train from scratch 

```python
# Train from scratch
log_dir = os.path.join('resnet', 'checkpoint')
accs = []
for i in range(4):
    acc = []
    log_file = os.path.join(log_dir, 'log_Res18_sp{}.txt'.format(i))
    with open(log_file) as lines:
        for line in lines:
            ep, a = line.strip().split(',')
            acc.append(float(a))
    accs.append(acc)
    if len(acc) > 1:
        print(np.max(np.asarray(acc)))
accs=np.asarray(accs)
```

```python
for i in range(4):
    plt.plot(accs[i])
plt.ylim([80,100])
plt.grid()
plt.show()
maxs=np.max(accs,axis=1)
print('logs: {}'.format(maxs))
print('mean: {}'.format(np.mean(maxs)))
print('std: {}'.format(np.std(maxs)))
```
